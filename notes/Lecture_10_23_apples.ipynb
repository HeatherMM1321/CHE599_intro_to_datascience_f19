# Histograms are not great for estimating deities 

*the underlying data generating dist'n is not jagged and discontinoue 
*these discontininuitites/drop/offs to zero depend heavily on the bin choice (centers, widths) 
*the density estimate depends on where the bins start and end 
*qualitative features on the underlying data generating dist'n can be hiddden depending on how the bins are chosen (including small shifts in the bins, even if they are the same width!!) 
*(the curse of dimensionality): as we add dimensions, the number of bins grows exponentially : thus most bins will be empyt in a higher dimensional histogram unless we have enrmous amout of data

```julia
bins = range(0.0, stop=130.0, length=8)

offsets = range(0.0, stop=50.0, length= 5)


for offset in offsets
    figure()
    xlabel(L"apples mass [g], $m$") #the L converts to latex
    scatter(df[:, :mass], [-0.001 for _ = 1:nrow(df)], marker="+",  color="r" )
    hist(df[:, :mass], bins=bins .+ offset)
    xlim([0, 170])
    ylabel(L"density $f(m)$")
end

#shifting bins can completley change the appereance of a histogram 
```
